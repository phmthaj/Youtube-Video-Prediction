```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
import datetime
import time
from datetime import date 
df = pd.read_csv('USvideos_filtered_with_durations_2.csv')
df.head()
```

```python
df_1 = pd.read_json("CA_category_id.json")
df_categories = pd.json_normalize(df_1['items'])
df_categories = df_categories[['id', 'snippet.title']]
df_categories.columns = ['Category ID', 'Category Title']
print(df_categories)
```

```python
def calculate_score(row):
    return (row['comment_count'] / row['views']) * \
           (row['likes'] - 1.5*row['dislikes'])

df['Score'] = df.apply(calculate_score, axis=1)
df.head()
```

```python
import re
def clean_description(description):
    description = re.sub(r'http[s]?://\S+|www\.\S+', '', description)
    description = re.sub(r'[^a-zA-Z0-9\s.,!?]', '', description)
    description = ' '.join(description.split())
    return description

df['description'] = df['description'].astype(str).apply(clean_description)
```

```python
def combine_text(row):
    return f"{row['title']} [SEP] {row['description']} [SEP] {row['tags']}"

df['text_combined'] = df.apply(combine_text, axis=1)
df.head()
```

```python
def classify_based_on_score(row):
    if row['views'] < 100000:
        return 0
    elif row['views'] >= 100000 and row['Score'] < 0:
        return 1
    elif row['views'] >= 100000 and 0 <= row['Score'] < 500:
        return 2
    elif row['views'] >= 100000 and row['Score'] >= 500:
        return 3

df['Y'] = df.apply(classify_based_on_score, axis=1)
df.head()
```

```python
from sentence_transformers import SentenceTransformer
from tensorflow.keras.models import load_model
model_bert = SentenceTransformer('all-MiniLM-L6-v2')
sentences = df['text_combined'].tolist()
embeddings = model_bert.encode(sentences, show_progress_bar=True)

df['text_embeddings'] = embeddings.tolist()
df.to_pickle('US_embedding_duration.pkl')
print("Đã lưu DataFrame vào file pickle: US_embedding_duration.pkl")
```

```python
from datetime import datetime

df["publish_time"] = pd.to_datetime(df["publish_time"].str.replace("Z", ""), format="%Y-%m-%dT%H:%M:%S.%f")
df["trending_date"] = pd.to_datetime(df["trending_date"], format="%y.%d.%m")
df["days_to_trending"] = (df["trending_date"] - df["publish_time"]).dt.days
df.head()
```

```python
df.to_pickle('US_embedding_duration_re.pkl')
print("Đã lưu DataFrame vào file pickle: US_embedding_duration_re.pkl")
```

```python
import joblib
df = pd.read_pickle('US_embedding_duration_re.pkl')

X_text_embeddings = np.stack(df['text_embeddings'].values)  
X_category_id = df['category_id'].values.reshape(-1, 1)     
X_days_to_trending = df['days_to_trending'].values.reshape(-1, 1)  
X_video_durations = df['video_durations'].values.reshape(-1, 1) 
y_true = df['Y'].values  

scaler_text = joblib.load('scaler_text_xgb.pkl')
scaler_category = joblib.load('scaler_category_xgb.pkl')
scaler_days = joblib.load('scaler_days_xgb.pkl')
scaler_duration = joblib.load('scaler_duration_xgb.pkl')

X_text_embeddings_scaled = scaler_text.transform(X_text_embeddings)
X_category_id_scaled = scaler_category.transform(X_category_id)
X_days_to_trending_scaled = scaler_days.transform(X_days_to_trending)
X_video_durations_scaled = scaler_duration.transform(X_video_durations)

X_combined = np.hstack([
    X_category_id_scaled,         
    X_days_to_trending_scaled,
    X_video_durations_scaled,    
    X_text_embeddings_scaled      
])

model = joblib.load('xgboost_model.pkl')

Y_predict = model.predict(X_combined)

df['Y_predict'] = Y_predict
df['Y'] = y_true

accuracy = (df['Y'] == df['Y_predict']).mean() * 100

print("DataFrame với Y và Y_predict:")
print(df[['Y', 'Y_predict']].head())  
print(f"\nTỷ lệ phần trăm Y_predict khớp với Y: {accuracy:.2f}%")

df.to_pickle('US_embedding_duration_with_predictions.pkl')
print("Đã lưu DataFrame với dự đoán vào file pickle: US_embedding_duration_with_predictions.pkl")
```
